# Transfert de style artistique avec TensorFlow Lite

Transfert de style artistique: créer une nouvelle image basée sur deux images d'entrée, l'une représentant le style artistique et l'autre représentant le contenu.

### Modèle utilisé

Ce modèle de transfert de style artistique se compose de deux sous-modèles:

1. **Modèle de prédiction de style :** un réseau de neurones basé sur MobilenetV2 qui prend une image de style d'entrée vers un vecteur de goulot d'étranglement de style à 100 dimensions.
2. **Modèle de transformation de style :** un réseau de neurones qui applique un vecteur de goulot d'étranglement de style à une image de contenu et crée une image stylisée.

![title](https://storage.googleapis.com/download.tensorflow.org/models/tflite/arbitrary_style_transfer/architecture.png)

```python
import tensorflow as tf
import IPython.display as display

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (12,12)
mpl.rcParams['axes.grid'] = False

import numpy as np
import time
import functools
```

On télécharge le contenu et les images de style, ainsi que les modèles TensorFlow Lite pré-entraînés.

```python
content_path = tf.keras.utils.get_file('carla.jpg', 'https://resize.programme-television.ladmedia.fr/r/670,670/img/var/premiere/storage/images/tele-7-jours/news-tv/carla-moreau-les-marseillais-apres-l-affaire-de-sorcellerie-elle-repond-a-une-nouvelle-polemique-4670722/99427758-1-fre-FR/Carla-Moreau-Les-Marseillais-Apres-l-affaire-de-sorcellerie-elle-repond-a-une-nouvelle-polemique.jpg')
style_path = tf.keras.utils.get_file('vangogh.jpg','https://www.kazoart.com/blog/wp-content/uploads/2017/03/van-gogh-la-nuit-etoilee-1.jpg')

style_predict_path = tf.keras.utils.get_file('style_predict.tflite', 'https://tfhub.dev/google/lite-model/magenta/arbitrary-image-stylization-v1-256/int8/prediction/1?lite-format=tflite')
style_transform_path = tf.keras.utils.get_file('style_transform.tflite', 'https://tfhub.dev/google/lite-model/magenta/arbitrary-image-stylization-v1-256/int8/transfer/1?lite-format=tflite')
```

## Pré-traitement des entrées

- L'image de contenu et l'image de style doivent être des images RVB dont les valeurs de pixel sont des nombres float32 compris entre [0..1].
- La taille de l'image de style doit être (1, 256, 256, 3). Nous recadrons l'image au centre et la redimensionnons.
- L'image de contenu doit être (1, 384, 384, 3). Nous recadrons l'image au centre et la redimensionnons.

```python
# Chargement d'une image avec un fichier, ajout d'une dimension 'batch'
def load_img(path_to_img):
    img = tf.io.read_file(path_to_img)
    img = tf.io.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]

    return img

# Pré-process de l'image : recadrer l'image centrale et redimensionnement
def preprocess_image(image, target_dim):
  # Redimensionnement de l'image de sorte que la dimension la plus courte devienne 256px
    shape = tf.cast(tf.shape(image)[1:-1], tf.float32)
    short_dim = min(shape)
    scale = target_dim / short_dim
    new_shape = tf.cast(shape * scale, tf.int32)
    image = tf.image.resize(image, new_shape)

    # Recadrage central de l'image
    image = tf.image.resize_with_crop_or_pad(image, target_dim, target_dim)

    return image

# Chargement des images d'entrées
content_image = load_img(content_path)
style_image = load_img(style_path)

# Pré-process de ces images
preprocessed_content_image = preprocess_image(content_image, 384)
preprocessed_style_image = preprocess_image(style_image, 256)

print('Style Image Shape:', preprocessed_style_image.shape)
print('Content Image Shape:', preprocessed_content_image.shape)
```

## Visualisation des entrées

```python
def imshow(image, title=None):
    if len(image.shape) > 3:
        image = tf.squeeze(image, axis=0)

    plt.imshow(image)
    if title:
        plt.title(title)

plt.subplot(1, 2, 1)
imshow(preprocessed_content_image, 'Content Image')

plt.subplot(1, 2, 2)
imshow(preprocessed_style_image, 'Style Image')
```

## Visualisation des entrées
### Prédiction de style 

```python
# Fonction permettant d'exécuter la prédiction du style sur une image de style prétraitée.
def run_style_predict(preprocessed_style_image):
    # Chargement du modèle
    interpreter = tf.lite.Interpreter(model_path=style_predict_path)

    # Fixer l'entrée du modèle
    interpreter.allocate_tensors()
    input_details = interpreter.get_input_details()
    interpreter.set_tensor(input_details[0]["index"], preprocessed_style_image)

    # Calculer le "goulot" de style
    interpreter.invoke()
    style_bottleneck = interpreter.tensor(
      interpreter.get_output_details()[0]["index"]
      )()

    return style_bottleneck

# Calculer le goulot d'étranglement du style pour l'image de style prétraitée.
style_bottleneck = run_style_predict(preprocessed_style_image)
print('Style Bottleneck Shape:', style_bottleneck.shape)
```

### Transformation de style 

```python
# Transformation de style sur l'image de style prétraitée
def run_style_transform(style_bottleneck, preprocessed_content_image):
    # Chargement du modèle
    interpreter = tf.lite.Interpreter(model_path=style_transform_path)

    # Fixer l'entrée du modèle
    input_details = interpreter.get_input_details()
    interpreter.allocate_tensors()

    # Fixer les entrées du modèle
    interpreter.set_tensor(input_details[0]["index"], preprocessed_content_image)
    interpreter.set_tensor(input_details[1]["index"], style_bottleneck)
    interpreter.invoke()

    # Transformer l'image de contenu
    stylized_image = interpreter.tensor(
      interpreter.get_output_details()[0]["index"]
      )()

    return stylized_image

# Styliser l'image de contenu en utilisant le goulot d'étranglement de style.
stylized_image = run_style_transform(style_bottleneck, preprocessed_content_image)

# Visualization de la sortie
imshow(stylized_image, 'Stylized Image')
```

### Mélange de styles
Nous pouvons mélanger le style de l'image de contenu dans la sortie stylisée, ce qui à son tour fait ressembler davantage la sortie à l'image de contenu.



```python
# Calcule du style bottleneck de l'image de contenu
style_bottleneck_content = run_style_predict(
    preprocess_image(content_image, 256)
    )
```

```python
# Définir le rapport de blending de contenu entre [0..1].
# 0.0: 0% de style extrait de l'image du contenu.
# 1.0: 100% de style extrait de l'image du contenu.
content_blending_ratio = 0.5

# Mélangez le goulot d'étranglement de l'image de style et de l'image de contenu.
style_bottleneck_blended = content_blending_ratio * style_bottleneck_content + (1 - content_blending_ratio) * style_bottleneck

# Styliser l'image de contenu en utilisant le goulot d'étranglement de style.
stylized_image_blended = run_style_transform(style_bottleneck_blended,
                                             preprocessed_content_image)

# Visualisation de la sortie
imshow(stylized_image_blended, 'Blended Stylized Image')
```
